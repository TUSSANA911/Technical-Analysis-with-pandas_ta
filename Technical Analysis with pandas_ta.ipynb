```python
import pandas_ta as ta
import yfinance as yf
import mplfinance as mpf
```

```python
data = yf.download('BTC-USD', interval='1d',start='2023-01-01', end='2023-12-20')
```

```python
data
```

```python
data.ta.indicators()
```

```python
data.ta.macd(fast=12,slow=21,signal=7)
```

```python
df_ta = data.copy()
```

```python
df_ta.ta.macd(slow=21,append=True)
```

```python
df_ta
```

```python
rsi = df_ta.ta.rsi()
```

```python
df_ta.ta.rsi(append=True)
df_ta
```

```python
df_ta.ta.ema(length=100,append=True)
df_ta
```

```python
df_ta.columns = ['Open',	'High',	'Low',	'Close',	'Adj Close',	'Volume',	'macd',	'histogram',	'signal_line',	'rsi']
```

```python
df_ta.columns 
```

```python
#for plot MACD
mpf_macd = mpf.make_addplot(df_ta['macd'],panel = 1,color = 'blue',ylabel='macd')
mpf_signal = mpf.make_addplot(df_ta['signal_line'],panel = 1,color = 'orange')

colors = ['green' if x >= 0 else 'red' for x in df_ta["histogram"]]
mpf_histogram = mpf.make_addplot(df_ta['histogram'],type='bar',panel = 1,color = colors)

mpf_rsi = mpf.make_addplot(df_ta['rsi'],panel=2,color='orange',ylabel='rsi')

mpf_ema100 = mpf.make_addplot(df_ta['EMA_100'],panel=0,color='orange')

ind_plot = [mpf_macd,mpf_signal,mpf_histogram,mpf_rsi,mpf_ema100]
```

```python
mpf.plot(df_ta,
         type='candle',
         style='charles',
         addplot=ind_plot,
         panel_ratios=(3,1,1),
         figratio=(2,1),
         figscale=1.5)
```

```python
mpf.plot(df_ta,
         type='candle',
         style='default',
         addplot=ind_plot, 
         hlines = dict(hlines=[40000, 30000, 20000],
         colors=['r','b','g'], 
         linestyle='-'))
```

```python
#Add line for RSI

df_ta['upper'] = 70
df_ta['lower'] = 30
```

```python
#for plot MACD
mpf_macd = mpf.make_addplot(df_ta['macd'],panel = 1,color = 'blue',ylabel='macd')
mpf_signal = mpf.make_addplot(df_ta['signal_line'],panel = 1,color = 'orange')

colors = ['green' if x >= 0 else 'red' for x in df_ta["histogram"]]
mpf_histogram = mpf.make_addplot(df_ta['histogram'],type='bar',panel = 1,color = colors)

mpf_rsi = mpf.make_addplot(df_ta['rsi'],panel=2,color='orange',ylabel='rsi')
mpf_rsi_upper = mpf.make_addplot(df_ta['upper'],panel=2,color='r')
mpf_rsi_lower = mpf.make_addplot(df_ta['lower'],panel=2,color='r')

mpf_ema100 = mpf.make_addplot(df_ta['EMA_100'],panel=0,color='orange')

ind_plot = [mpf_macd,mpf_signal,mpf_histogram,mpf_rsi,mpf_rsi_upper,mpf_rsi_lower,mpf_ema100]
```

```python
mpf.plot(df_ta,
         type='candle',
         style='yahoo',
         addplot=ind_plot,
         panel_ratios=(3,1,1),
         figratio=(2,1.5),
         hlines = dict(hlines=[30000, 25000])
        )
```
